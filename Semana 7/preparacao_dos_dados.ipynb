4. Após aplicação das operações, faça a representação gráfica com histogramas das novas características dos dados analisados.

```python
import numpy as np

df_nan = df.copy()
n = max(1, int(len(df_nan) * 0.5))
df_nan.loc[:n-1, "residual sugar"] = np.nan
```

```python
media = df_nan["residual sugar"].mean()
df_nan["residual sugar"] = df_nan["residual sugar"].fillna(media)
df_nan
```

```python
fig, axes_1 = plt.subplots(1, 2, figsize=(10,5))

axes_1[0].set_title("Residual Sugar Normal")
axes_1[0].hist(df['residual sugar'])

axes_1[1].set_title("Residual Sugar NaN")
axes_1[1].hist(df_nan['residual sugar'])
```

6. No dataset escolhido selecione intervalos de amostragem (10%, 30% e 50% - Progressiva) e faça análises sobre os dados em relação as métricas estatísticas. O que você pode observar sobre os subconjuntos em relação as métricas estatísticas com a nova composição dos dados?

```python
df_10 = df.iloc[:int(len(df) * 0.10)]
df_30 = df.iloc[:int(len(df) * 0.30)]
df_50 = df.iloc[:int(len(df) * 0.50)]
```

```python
print("10%")
print("Média:\n",    df_10.mean().round(3))
print("Mediana:\n",  df_10.median().round(3))
print("Desvio padrão:\n", df_10.std().round(3))
```

```python
print("30%")
print("Média:\n",    df_30.mean().round(3))
print("Mediana:\n",  df_30.median().round(3))
print("Desvio padrão:\n", df_30.std().round(3))
```

```python
print("50%")
print("Média:\n",    df_50.mean().round(3))
print("Mediana:\n",  df_50.median().round(3))
print("Desvio padrão:\n", df_50.std().round(3))
```

```python
print("100%")
print("Média:\n",    df.mean().round(3))
print("Mediana:\n",  df.median().round(3))
print("Desvio padrão:\n", df.std().round(3))
```

Foi observado que com poucos dados as métricas tornam-se instáveis e pouco confiáveis, pois dependem dos registros que caem na amostra. Com 50% os valores já estavam próximos do real, podendo servir como uma boa base de análise. Com 10%, embora alguns atributos tenham chegado perto por coincidência, não há garantia que estejam representando os dados em sua totalidade.
